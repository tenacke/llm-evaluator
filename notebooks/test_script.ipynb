```python
import ollama 
import os
import pandas as pd
```

```python
model_name = 'evallm:v2'
number_of_repets = 3
```

```python
powerful_df = pd.read_csv('csv/powerful_model.csv')
powerful_df = powerful_df.drop(columns=['Unnamed: 0'])
powerful_df.head(3)
```

```python
client = ollama.Client()
client
```

```python
results_file_name = 'model2results.csv'
results_file = open(results_file_name, 'a')
results_file.write('index,rep,response\n')
```

```python
for row in powerful_df.iterrows():
    index, data = row
    file = data['filepath']
    summary = data['decoded']

    print(f'Processing summary {index}...')

    with open(os.path.join('datasets', file), 'r') as f:
        text = f.read()

    query = 'Text:\n' + text + 'Summary:\n' + summary

    result = {
        'coherence': 0,
        'fluency': 0,
        'relevance': 0,
        'consistency': 0
    }
    for i in range(number_of_repets):
        print(f'Repetition {i+1}...')
        response = client.generate(model_name, query)
        response_str = response.response

        results_file.write(f'{index},{i},{response_str}\n')


    print('Repets are done')
```

```python
results_file.close()
```
