```python
import ollama 
import os
import pandas as pd
```

```python
model_name = 'evallm-consistency:v2.1'
number_of_repets = 3
```

```python
powerful_df = pd.read_csv('csv/average_model.csv')
powerful_df = powerful_df.drop(columns=['Unnamed: 0'])
powerful_df.head(3)
```

```python
client = ollama.Client()
client
```

```python
deneme_df = powerful_df.head(5)
```

```python
results = pd.DataFrame(columns=['result'], index = range(0, len(deneme_df)))
results.head(3)
```

```python
for row in powerful_df.iterrows():
    index, data = row
    file = data['filepath']
    summary = data['decoded']

    print(f'Processing summary {index}...')

    with open(os.path.join('datasets', file), 'r') as f:
        text = f.read()

    query = 'Text:\n' + text + 'Summary:\n' + summary

    result = 0
    for i in range(number_of_repets):
        print(f'Repetition {i+1}...')
        response = client.generate(model_name, query)
        response_str = response.response

        point = int(response_str.split('Score: ')[1][0])

        result += point

    print('Repets are done')
    result = result / number_of_repets
    results.loc[index] = {'result': result}
```

```python
results
```

```python
# write the results to csv
results.to_csv('csv/ollama_results_consistency_average.csv')
```
