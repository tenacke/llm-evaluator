```python
import os
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
dir_path = os.path.join("datasets", "cnndm")
```

```python
# Path to the JSONL file
file_path = os.path.join(dir_path, "model_annotations.aligned.jsonl")

# Read and parse the JSONL file
with open(file_path, "r", encoding="utf-8") as file:
    data = [json.loads(line) for line in file]

# create pandas dataframe
df = pd.DataFrame(data)
df.head()
```

```python
df = df[['model_id', 'filepath', 'decoded', 'expert_annotations', 'turker_annotations', 'references']]
df.head()
```

```python
expert_annotations = df["expert_annotations"]
turker_annotations = df["turker_annotations"]
```

```python
# for each row in the expert_annotations column, get the array of annotations and extract coherence score
expert_coherence_scores = expert_annotations.apply(lambda x: [annotation["coherence"] for annotation in x])

# for each row in the turker_annotations column, get the array of annotations and extract coherence score
turker_coherence_scores = turker_annotations.apply(lambda x: [annotation["coherence"] for annotation in x])

# get the average coherence score for each row
expert_coherence_scores = expert_coherence_scores.apply(lambda x: sum(x) / len(x))
turker_coherence_scores = turker_coherence_scores.apply(lambda x: sum(x) / len(x))

# add the average coherence scores to the dataframe
df["expert_coherence_score"] = expert_coherence_scores
df["turker_coherence_score"] = turker_coherence_scores
```

```python
# for each row in the expert_annotations column, get the array of annotations and extract relevance score
expert_relevance_scores = expert_annotations.apply(lambda x: [annotation["relevance"] for annotation in x])

# for each row in the turker_annotations column, get the array of annotations and extract relevance score
turker_relevance_scores = turker_annotations.apply(lambda x: [annotation["relevance"] for annotation in x])

# get the average relevance score for each row
expert_relevance_scores = expert_relevance_scores.apply(lambda x: sum(x) / len(x))
turker_relevance_scores = turker_relevance_scores.apply(lambda x: sum(x) / len(x))

# add the average relevance scores to the dataframe
df["expert_relevance_score"] = expert_relevance_scores
df["turker_relevance_score"] = turker_relevance_scores
```

```python
# for each row in the expert_annotations column, get the array of annotations and extract consistency score
expert_consistency_scores = expert_annotations.apply(lambda x: [annotation["consistency"] for annotation in x])

# for each row in the turker_annotations column, get the array of annotations and extract consistency score
turker_consistency_scores = turker_annotations.apply(lambda x: [annotation["consistency"] for annotation in x])

# get the average consistency score for each row
expert_consistency_scores = expert_consistency_scores.apply(lambda x: sum(x) / len(x))
turker_consistency_scores = turker_consistency_scores.apply(lambda x: sum(x) / len(x))

# add the average consistency scores to the dataframe
df["expert_consistency_score"] = expert_consistency_scores
df["turker_consistency_score"] = turker_consistency_scores
```

```python
# for each row in the expert_annotations column, get the array of annotations and extract fluency score
expert_fluency_scores = expert_annotations.apply(lambda x: [annotation["fluency"] for annotation in x])

# for each row in the turker_annotations column, get the array of annotations and extract fluency score
turker_fluency_scores = turker_annotations.apply(lambda x: [annotation["fluency"] for annotation in x])

# get the average fluency score for each row
expert_fluency_scores = expert_fluency_scores.apply(lambda x: sum(x) / len(x))
turker_fluency_scores = turker_fluency_scores.apply(lambda x: sum(x) / len(x))

# add the average fluency scores to the dataframe
df["expert_fluency_score"] = expert_fluency_scores
df["turker_fluency_score"] = turker_fluency_scores
```

```python
df.drop(columns=["expert_annotations", "turker_annotations"], inplace=True)

df.head()
```

```python
df.sort_values(by='model_id', inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
numeric_df = df.drop(columns=["filepath", "decoded", "references"])
numeric_df.head()
```

```python
average_scores = numeric_df.groupby("model_id").mean()
average_scores
```

```python
powerful_model = 'M22'
poor_model = 'M11'
average_model = 'M8'
```

```python
powerful_df = df[df["model_id"] == powerful_model]
poor_df = df[df["model_id"] == poor_model]  
average_df = df[df["model_id"] == average_model]

powerful_df.head()
```

```python
powerful_df.to_csv(os.path.join(dir_path, "powerful_model.csv"))
poor_df.to_csv(os.path.join(dir_path, "poor_model.csv"))
average_df.to_csv(os.path.join(dir_path, "average_model.csv"))
```
